```python
Дана функция f(x) = 5x^2 + 10x - 30

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0
```

```python
from sympy import *
x = Symbol('x')
print(solve(5*x**2 + 10*x - 30))
```

```python
from sympy import *
x = Symbol('x')
f = 5 * x ** 2 + 10 * x - 30
dir_f = solve(f.diff(x))
dir_y = (5*dir_f[0]**2 + 10*dir_f[0]-30)
print(f'Функция убывает на промежутке (-∞;{dir_f[0]}), функция возрастает на промежутке ({dir_f[0]}; +∞)Вершина является({dir_f[0]};{dir_y})')
```
